```python
from PyPDF2 import PdfFileReader
def extract_pdfText():
    with open(r'C:\Users\Python_learning\\sustainability.pdf','rb') as f:
        pdf = PdfFileReader(f)
        # get the 50th page
        page = pdf.getPage(1)
        print(page)
        print('Page type: {}'.format(str(type(page))))
        #Extract text from the 50th page
        text = page.extractText()
        print(text)
        if __name__ == '__main__':
            path = 'C:\Users\Python_learning\\sustainability.pdf'
            extract_pdfText(path)
```
